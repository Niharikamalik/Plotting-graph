```python
# plot funtion x^3
import numpy as np
x =  np.array([1,2,3,4])
y = x**3
plt.plot(x,y)
plt.show()
```

```python
x =  np.arange(0,4,0.1)
y = x**3
plt.plot(x,y,"go")
plt.show()
```

#### color = name of the color
#### marker = give shapes
#### linewidth = width of connecting line
#### label = define connected line belong to which function
#### legend() = used to show labels in graph 
#### ylabel () = label  y axis
#### xlable() = label x axis
#### title() = given title of graph 
#### axis( ) = axis([value for x axis , value for y axis])

```python
x =  np.arange(0,4,0.1)
y = x**3
y2 = x**2

plt.plot(x,y,color = 'red',marker="+", linewidth = 1 , label = "x^3")
plt.plot(x,y2,color = 'black',marker = "*" ,linewidth = 1 , label = "x^2")

plt.ylabel('y')
plt.xlabel('x')
plt.title('matplotlib Graph')
plt.legend()
plt.show()
```

```python
x =  np.arange(0,4,0.1)
y = x**3
y2 = x**2
plt.plot(x,y,color = 'red',marker="+", linewidth = 1 , label = "x^3")
plt.plot(x,y2,color = 'black',marker = "*" ,linewidth = 1 , label = "x^2")
plt.axis([1,10,1,100])
plt.grid()
plt.text(1.5,94,"text",fontsize = 14)
plt.ylabel('y')
plt.xlabel('x')
plt.title('matplotlib Graph')
plt.legend()
plt.show()
```

```python
# bubble chart 
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,15,16,17]
population = [100,200,300,400,500,600]
plt.scatter(year,salary,population,color = "black")
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,population,c= c  )
plt.title("Bubble chart")
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,population,c= c  )
plt.title("Bubble chart")
for i in range(len(year)):
    plt.text(year[i],salary[i] +0.3,population[i])
plt.show()
```

```python
#pie chart
a = [2,3,4,5]
colors = ["lightcyan","lightskyblue","lightpink","violet"]
plt.pie(a,colors=colors)
plt.show()
```

```python
a = [2,3,4,5]
labels = ["A","B","C","D"]
explode = [0,0.11,0,0]
colors = ["lightcyan","lightskyblue","lightpink","violet"]
plt.pie(a,colors=colors,explode=explode ,labels=labels,autopct="%.1f%%",counterclock=False,startangle=102,shadow = True)
plt.axis('equal')
plt.show()
```
